```python
df['ORDERDATE']
df['DATE'] = pd.to_datetime(df['ORDERDATE'])
df['DATE']


df['DaysOfWeek'] = df['DATE'].dt.day_name()
df['DaysOfWeek'] 

group_by_DaysOfWeek = df.groupby('DaysOfWeek')['SALES'].sum()
group_by_DaysOfWeek

Days_Of_Week = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']

group_by_DaysOfWeek =group_by_DaysOfWeek.reindex(Days_Of_Week)
group_by_DaysOfWeek

group_by_DaysOfWeek.plot()
```

```python
df['DATE'] =pd.to_datetime( df['ORDERDATE'])

df['MONTHSOFYEAR'] = df['DATE'].dt.month_name()
df['MONTHSOFYEAR']

group_by_MonthOfYear = df.groupby('MONTHSOFYEAR')['SALES'].sum()
group_by_MonthOfYear

Months_of_year = ['January', 'February' ,'March', 'April','May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']

group_by_MonthOfYear = group_by_MonthOfYear.reindex(Months_of_year)
group_by_MonthOfYear

group_by_MonthOfYear.plot()
```

```python
df['DAYSOFMONTH'] =df['DATE'].dt.day
df['DAYSOFMONTH']

sales_by_group_DaysOfMonth= df.groupby('DAYSOFMONTH')['SALES'].sum()
sales_by_group_DaysOfMonth

sales_by_group_DaysOfMonth.plot()
```
